# CNN 개요

합성곱 신경망(Convolutional Neural Network, CNN)은 주로 이미지와 같은 2차원 데이터 분석에 적합한 신경망 구조이다.

합성곱과 풀링 계층을 사용하여 특징을 추출하고, 최종적으로 완전 연결 계층을 통해 분류 작업을 수행한다.

CNN은 이미지의 공간적 구조를 유지하면서 각 픽셀 간의 관계를 고려할 수 있는 특성이 있어, 이미지 인식, 영상 분석, 그리고 자연어 처리에서도 널리 활용된다.

1. 합성곱 층(Convolutional Layer)
    - **합성곱 연산**을 통해 이미지의 중요한 특징을 추출한다. 작은 필터(또는 커널)를 사용하여 이미지 전체를 슬라이딩하며 각 영역의 특징을 추출한다.
    - 필터는 각각 특정 패턴(예: 가장자리, 텍스처 등)을 감지하는데, 이 필터들을 학습하면서 이미지의 중요한 정보가 추출된다.
    - 각 필터의 이동 결과를 **특징 맵(feature map)**이라고 하며, 이 특징 맵들이 합쳐져서 이미지의 중요한 정보를 단계적으로 추출한다.

    - **활성화 함수**
        - 합성곱 층의 출력을 비선형 함수인 **ReLU** 함수에 통과시켜 비선형성을 부여한다. 이를 통해 네트워크가 더 복잡한 패턴을 학습할 수 있게 된다.

    - 풀링 층(Pooling Layer)
        - **맥스 풀링(Max Pooling)**이나 **평균 풀링(Average Pooling)**과 같은 연산을 통해 특징 맵의 크기를 줄여 데이터의 차원을 줄이고 계산량을 줄인다.
        - 일반적으로 가장 높은 값을 취하는 맥스 풀링이 자주 사용되며, 이를 통해 이미지의 주요 특징이 강조된다.
        - 풀링은 데이터의 위치에 대한 불변성을 높이는 데 도움을 주며, 과적합을 방지하는 효과도 있다.

2. 완전 연결 층(Fully Connected Layer)
    - 마지막으로, 추출된 특징을 활용하여 분류 작업을 수행하기 위해 **완전 연결 층**을 사용한다.
    - CNN의 마지막 단계에서 1차원 벡터 형태로 평탄화(Flatten)된 후, 완전 연결 층으로 연결되어 각 클래스에 대한 확률값을 출력한다.
  
3. CNN의 학습
    - CNN은 데이터의 특징을 학습하며 필터와 가중치를 최적화해나간다. 이를 위해 **역전파(backpropagation)**와 **경사하강법(gradient descent)**을 사용하여 필터의 파라미터를 갱신한다.
    - CNN은 이미지에서 가장자리, 질감, 모양 등의 낮은 수준의 특징을 추출한 후 점차 높은 수준의 의미 있는 특징을 추출해 최종적으로 분류를 가능하게 한다.

## Convolution Layer 주요개념

**Feature Extraction**: 이미지의 특정 패턴이나 특징을 추출하는 과정으로, CNN에서 필터와 합성곱 연산을 통해 이미지의 중요한 특징을 뽑아낸다.

**Filter (Kernel)**: 작은 행렬로, 이미지의 특정 패턴을 인식하는 역할을 한다. 필터를 이미지에 적용해 특징 맵(feature map)을 생성한다.

- **Valid Padding**: 패딩을 주지 않는 방식으로, 출력 크기는 입력보다 작아진다.(기본값)
- **Same Padding**: 출력 크기가 입력과 동일하도록 패딩을 적용한다.

## conv layer + pooling layer

```python
# Pytorch CNN 기본 흐름: Conv2d -> ReLU -> MaxPool2d 
import torch
import torch.nn as nn 
import torch.nn.functional as F 

# 1. 입력 텐서 생성
input_tensor = torch.randn(1, 1, 28, 28) # (배치 크기=1, 채널=1[흑백], 높이=28, 너비=28)

# 2. 합성곱 층(Convolution Layer) 정의
conv_layer = nn.Conv2d(
    in_channels = 1,      # 입력 채널 수 (흑백 이미지)
    out_channels = 4,     # 출력 채널 수 (4개의 특징 맵 생성)
    kernel_size = 3,      # 커널(필터) 크기 3x3
    stride=1,             # 스트라이드 1 (1칸씩 이동)
    padding='same'        # 패딩 'same' (입력과 출력 크기 동일 유지)
)

# 3. 합성곱 연산 및 활성화 함수 적용
x = conv_layer(input_tensor) # 합성곱 적용
x = F.relu(x)                # ReLU 활성화 함수 적용 (비선형성 추가)
print(f'Conv2d & ReLU Output Shape: {x.shape}') # 출력: [1, 4, 28, 28]

# 4. 풀링 층(Pooling Layer) 적용
x = F.max_pool2d(x, kernel_size=2, stride=2) # 2x2 영역에서 최대값 추출, 크기 1/2로 축소
print(f'MaxPool2d Output Shape: {x.shape}')  # 출력: [1, 4, 14, 14]
```

```python
# 이미지 전처리 : 평균 0, 표준편차 1범위로 변환
train_images = train_dataset.data.unsqueeze(1) / 255.0  # (N, 28, 28) ->  (N, 1, 28, 28) 채널추가 + 0~1 변환
train_images = (train_images - 0.5) / 0.5               # 0 ~ 1 값 분포를 -1~1로 변환


test_images = test_dataset.data.unsqueeze(1) / 255.0
test_images = (test_images - 0.5) / 0.5

train_labels = train_dataset.targets    # 학습 라벨 텐서
test_labels = test_dataset.targets

print(train_images.shape)       # 전처리된 학습 이미지 shape
train_images[0][0]              # 첫번째 이미지의 0번째 채널값 확인
```

train_images = (train_images - 0.5) / 0.5
- 이미지가 0~1 범위에 있고, 평균이 대략적으로 0.3~0.6 근처라서 0.5 기준으로 이동을 시키면 대략적으로 평균 0 근처로 맞춰지는 효과가있다.
- CNN에서 -1~1 범위로 맞추는 것은 학습 안정성이 좋아지는 효과가 있다. (빠른 안정성을 위한 범위 스케일링)

```python
from sklearn.model_selection import train_test_split

tr_idx, val_idx = train_test_split(
    range(len(train_images)),       # 전체 인덱스
    test_size=0.15,                 # 15% 검증용
    random_state=42                 # 재현성 시드고정
)

tr_images = train_images[tr_idx]    # 학습 인덱스에 해당하는 이미지
tr_labels = train_labels[tr_idx]    # 학습 인덱스에 해당하는 라벨

val_images = train_images[val_idx]   #
val_labels = train_labels[val_idx]

print(tr_images.shape, tr_labels.shape)
print(val_images.shape, val_labels.shape)
print(test_images.shape, test_labels.shape)
```

```python
# FashionNMIST 분류용 CNN 모델 정의
import torch
import torch.nn as nn 
import torch.optim as optim
import torch.nn.functional as F
from torchsummary import summary

# FashionNMIST 분류용 CNN 모델 클래스
class FashionMNISTCNN(nn.Module):
    def __init__(self):
        super().__init__()

        # Convolution Layer (Feature Extraction 특징 추출)
        self.conv1 = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(64, 32, kernel_size=3, stride=1, padding=0)
        self.pool = nn.MaxPool2d(kernel_size = 2, stride=2)                     

        # Fully Connected Layer (분류기)
        # - 3차원으로 되어있는 Feature Map 결과를 Fully Connected 연결하기 위해서 Flatten()을 적용해야 한다.
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(32 * 13 * 13, 100)
        self.fc2 = nn.Linear(100, 10)
    
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = self.pool(x)

        x = self.flatten(x)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = FashionMNISTCNN()
summary(model, (1, 28, 28), device = 'cpu')
```

```python

# Pytorch CNN 학습 파이프라인 : Device 설정 -> DataLoader -> Train/Validation Loop -> History 기록
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader   # 텐서 데이터셋 / 배치 로더

# 디바이스 / 손실함수 / 최적화함수 설정
device = torch.device('mps' if torch.mps.is_available() else 'cpu')

model = model.to(device)
criterion = nn.CrossEntropyLoss()    # 다중분류 손실함수 (로짓 입력 + 정수 라벨)
optimizer = optim.AdamW(model.parameters(), lr = 0.001)

# 데이터로더 생성
batch_size = 128
train_dataset = TensorDataset(tr_images, tr_labels)     # (이미지, 라벨) 텐서 묶기
val_dataset = TensorDataset(val_images, val_labels)

train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle=True)    # 학습 로더는 섞어서 배치 생성
val_loader = DataLoader(val_dataset, batch_size = batch_size, shuffle=False)        # 검증 로더는 고정

history = {'loss' : [], 'accuracy' : [], 'val_loss' : [], 'val_accuracy' : []}

epochs = 30
for epoch in range(epochs):
    # 학습 모드
    model.train()    # 드롭아웃, 배치 정규화 활용
    train_loss = 0.0
    train_correct = 0
    train_total = 0

    for batch_idxm, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)

        optimizer.zero_grad()   # 이전 배치 grafient 초기화
        outputs = model(data)   # 순전파 (로짓 출력)
        loss = criterion(outputs, target)   # 손실 계산
        loss.backward()     # 역전파(gradient 계산)
        optimizer.step()    # 파라미터 업데이트

        train_loss += loss.item()    # 배치 손실 누적 (스칼라)
        _, predicted = torch.max(outputs.data, 1)    # 가장 큰 로짓의 클래스 인덱스
        train_total += target.size(0)    # 배치 샘플 수 누적
        train_correct += (predicted == target).sum().item()   # 맞춘 개수 누적

    # 검증 모드
    model.eval()      # 드롭아웃, 배치 정규화 비활성화
    val_loss = 0.0
    val_correct = 0
    val_total = 0

    with torch.no_grad():    # 검증 시 gradient 계산 비활성화 (속도 / 메모리 절약)
        for data, target in val_loader:
            data, target = data.to(device), target.to(device)

            outputs = model(data)       # 순전파
            loss = criterion(outputs, target)    # 손실 계산

            val_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            val_total += target.size(0)
            val_correct += (predicted == target).sum().item()

    # 평균 계산
    train_loss /= len(train_loader)
    val_loss /= len(val_loader)
    train_accuracy = train_correct / train_total
    val_accuracy = val_correct / val_total

    history['loss'].append(train_loss)
    history['accuracy'].append(train_accuracy)
    history['val_loss'].append(val_loss)
    history['val_accuracy'].append(val_accuracy)

    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch+1}/{epochs}]')
        print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy:.4f}')
        print(f'Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy:.4f}')
        print('-'*50)

print('학습 완료!')
```

```python
# 학습 곡선 시각화 함수
import matplotlib.pyplot as plt

def show_history(history):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history['loss'], label='train')
    plt.plot(history['val_loss'], label='validation')
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['accuracy'], label='train')
    plt.plot(history['val_accuracy'], label='validation')
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

show_history(history)
```

```python
# 테스트 데이터로 평가 
test_dataset = TensorDataset(test_images, test_labels) 
test_loader = DataLoader(test_dataset, batch_size = 256, shuffle=False)   


model.eval()      # 드롭아웃, 배치 정규화 비활성화
test_loss = 0.0
test_correct = 0
test_total = 0

with torch.no_grad():    # 검증 시 gradient 계산 비활성화 (속도 / 메모리 절약)
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)

        outputs = model(data)       # 순전파
        loss = criterion(outputs, target)    # 손실 계산

        test_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        test_total += target.size(0)
        test_correct += (predicted == target).sum().item()

# 평균 계산
test_loss /= len(test_loader)
test_accuracy = test_correct / test_total

print(f'Test Loss: {test_loss:.4f}')
print(f'Test Acc: {test_accuracy:.4f}')
```
