```python
from google.colab import userdata
import os 

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
```

```python
# OpenAI Responses API로 이미지 생성 후 Base64 인코딩하여 파일로 저장
from openai import OpenAI  # OpenAI 라이브러리에서 OpenAI 클래스를 가져옴
import base64  # base64 인코딩 및 디코딩을 위한 모듈 가져옴
from io import BytesIO  # 바이트 스트림 처리를 위한 BytesIO 가져옴
client = OpenAI()  # OpenAI 클라이언트 인스턴스 생성

# Images API 호출
response = client.response.create(  # 이미지 생성을 위한 API 요청
    model = 'gpt-image-1',  # 사용할 모델을 'gpt-image-1'로 설정
    input = 'Draw a gorgeous image of a river made of white owl features, snaking its way though a serene winter landscape',  # 이미지 생성에 사용할 입력 문구
    size = '1024x1024',  # 생성할 이미지의 크기 설정
    quality = 'high',  # 이미지 품질 설정
    output_format = 'png'  # 출력 형식을 PNG로 설정
)

b64 = response.data[0].b64_json  # 응답에서 Base64 인코딩된 이미지 데이터 추출
img_bytes = base64.b64decode(b64)  # Base64 문자열을 바이트로 디코드

img = Image.open(BytesIO(img_bytes))  # 이미지 데이터를 열어 이미지 객체 생성
display(img)  # 생성된 이미지를 출력

with open('river.png', 'wb') as f:  # 'river.png'라는 이름으로 파일 열기
    f.write(img_bytes)  # 바이트 데이터를 파일에 쓰기
```

```python
def generate_image(prompt : str):
    response = client.response.create(  # 이미지 생성을 위한 API 요청
    model = 'gpt-image-1-mini',  # 사용할 모델을 'gpt-image-1'로 설정
    input = 'Draw a gorgeous image of a river made of white owl features, snaking its way though a serene winter landscape',  # 이미지 생성에 사용할 입력 문구
    size = '1024x1024',  # 생성할 이미지의 크기 설정
    quality = 'high',  # 이미지 품질 설정
    output_format = 'png'  # 출력 형식을 PNG로 설정
    )

    b64 = response.data[0].b64_json       # 응답에서 Base64 인코딩된 이미지 데이터 추출
    img_bytes = base64.b64decode(b64)     # Base64 문자열을 바이트로 디코드

    img = Image.open(BytesIO(img_bytes))  # 이미지 데이터를 열어 이미지 객체 생성
    display(img)  # 생성된 이미지를 출력
    
generate_image('프로젝트로 인해서 고통받고 있는 우리팀')
```
